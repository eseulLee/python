```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 📌 Line Plot

```python
fig, ax = plt.subplots()  # subplot() 괄호 안에 아무것도 없으면 1개의 그래프 생성
x = np.arange(15)
y = x ** 2
ax.plot(
    x, y,
    linestyle = ':',
    marker = '*',
    color = '#524FA1'
)
```

### 💡 Line Style

```python
x = np.arange(10)
fig, ax = plt.subplots()

# solid
ax.plot(x, x, linestyle = '-')

# dashed
ax.plot(x, x+2, linestyle = '--')

# dastdot
ax.plot(x, x+4, linestyle = '-.')

# dotted
ax.plot(x, x+6, linestyle = ':')
```

### 💡 Line Color

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, color = 'r')  # red
ax.plot(x, x+2, color = 'green')
ax.plot(x, x+4, color = '0.8')  # 0~1 사이의 grayscale
ax.plot(x, x+6, color = '#524FA1')
```

### 💡 Marker



```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, marker = '.')
ax.plot(x, x+2, marker = 'o')
ax.plot(x, x+4, marker = 'v')
ax.plot(x, x+6, marker = 's')  # square
ax.plot(x, x+8, marker = '*')
```

### 💡 축 경계 조정하기

```python
x = np.linspace(0, 10, 1000)  #  np.linspace(start, end, step)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-2, 12)    # x limit
ax.set_ylim(-1.5, 1.5) # y limit
```

### 💡 범례

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, label = 'y=x')
ax.plot(x, x**2, label = 'y=x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='upper right',
         shadow=True,
         fancybox=True,
         borderpad=2)
```

## 📌 Bar Plot

```python
x = np.arange(10)
fig, ax = plt.subplots(figsize=(12, 4))  # (가로, 세로)
ax.bar(x, x*2)  # (x, y)
```

```python
x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)
data = [x, y, z]

fig, ax = plt.subplots()
x_ax = np.arange(3)
for i in x_ax :
    ax.bar(x_ax, data[i],
    bottom = np.sum(data[:i], axis=0))

ax.set_xticks(x_ax)
ax.set_xticklabels(['A', "B", "C"])
```

## 📌 Histogram (도수분포표)

```python
fig, ax = plt.subplots()
data = np.random.randn(1000)
ax.hist(data, bins=50)   # bins : 막대의 개수
```

## 🥸 범례(legend) 위치 변경


```python
# legend 위치를 "왼쪽 중간"으로 세팅해보기
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(
    x, x, label = 'y=x',
    linestyle = '-',
    marker = '.',
    color = 'blue'
)
ax.plot(
    x, x**2, label= 'y=x^2',
    linestyle = '-.', 
    marker = ',',
    color = 'red'
)
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.legend(
    loc = 'center left',  # 'upper left'
    shadow=True,
    fancybox=True,
    borderpad=2
)
```

## 🥸 그래프 두개 그려보기!

```python
x = np.array(['축구','야구','농구','배드민턴', '탁구'])
y = np.array([13, 10, 17, 8, 7])
z = np.random.rand(1000)

fig, axes = plt.subplots(1, 2, figsize = (12, 4))

# Bar plot
axes[0].bar(x, y)

# HIstogram
axes[1].hist(z, bins=200)
```
