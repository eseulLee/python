```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
df.head()
```

### ü•∏ Question 1
ÎßàÏºÄÌåÖ ÏùëÎãµ Í≥†Í∞ùÎì§Ïùò ÎÇòÏù¥Î•º 10ÏÇ¥ Îã®ÏúÑÎ°ú Î≥ÄÌôò ÌñàÏùÑ Îïå, Í∞ÄÏû• ÎßéÏùÄ Ïù∏ÏõêÏùÑ Í∞ÄÏßÑ ÎÇòÏù¥ÎåÄÎäî? (0~9 : 0 , 10~19 : 10)

```python
def changeAge(x) :
    new_age = (x // 10) * 10
    return new_age

result1 = df.age.agg(changeAge).value_counts().index[0]
print(result1)
```

### ü•∏ Question 2
ÎßàÏºÄÌåÖ ÏùëÎãµ Í≥†Í∞ùÎì§Ïùò ÎÇòÏù¥Î•º 10ÏÇ¥ Îã®ÏúÑÎ°ú Î≥ÄÌôò ÌñàÏùÑ Îïå, Í∞ÄÏû• ÎßéÏùÄ ÎÇòÏù¥ÎåÄ Íµ¨Í∞ÑÏùò Ïù∏ÏõêÏùÄ Î™áÎ™ÖÏù∏Í∞Ä?

```python
result2 = df.age.agg(changeAge).value_counts().values[0]
print(result2)
```

### ü•∏ Question 3
ÎÇòÏù¥Í∞Ä 25ÏÇ¥ Ïù¥ÏÉÅ 29ÏÇ¥ ÎØ∏ÎßåÏù∏ ÏùëÎãµ Í≥†Í∞ùÎì§Ï§ë housingÏª¨ÎüºÏùò Í∞íÏù¥ yesÏù∏ Í≥†Í∞ùÏùò ÏàòÎäî?

```python
result3 = df[(df.age >= 25) & (df.age < 29) & (df.housing == 'yes')].shape[0]
print(result3)
```

### ü•∏ Question 4
numericÌïú Í∞íÏùÑ Í∞ÄÏßÄÏßÄ ÏïäÏùÄ Ïª¨ÎüºÎì§Ï§ë uniqueÌïú Í∞íÏùÑ Í∞ÄÏû• ÎßéÏù¥ Í∞ÄÏßÄÎäî Ïª¨ÎüºÏùÄ?

```python
cols = df.select_dtypes(exclude=int).columns
lst  = []
for col in cols :
    target = df[col]
    lst.append([col, target.nunique()])
result = pd.DataFrame(lst).sort_values(1, ascending=False).values[0][0]
print(result)
```

### ü•∏ Question 5
balance Ïª¨ÎüºÍ∞íÎì§Ïùò ÌèâÍ∑†Í∞í Ïù¥ÏÉÅÏùÑ Í∞ÄÏßÄÎäî Îç∞Ïù¥ÌÑ∞Î•º IDÍ∞íÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨ÌñàÏùÑÎïå ÏÉÅÏúÑ 100Í∞ú Îç∞Ïù¥ÌÑ∞Ïùò balanceÍ∞íÏùò ÌèâÍ∑†ÏùÄ?

```python
target = df[df.balance >= df.balance.mean()].sort_values(by='ID', ascending=False).head(100)
print(target.balance.mean())
```

### ü•∏ Question 9
y Î≥ÄÏàòÏôÄ education Î≥ÄÏàòÎäî ÎèÖÎ¶ΩÏù∏ÏßÄ Ïπ¥Ïù¥Ï†úÍ≥±Í≤ÄÏ†ïÏùÑ ÌÜµÌï¥ ÌôïÏù∏ÌïòÎ†§ÌïúÎã§. p-valueÍ∞íÏùÑ Ï∂úÎ†•ÌïòÎùº

```python
v = pd.crosstab(df.y, df.education)
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(v)

display(v)
print(p)
```

### ü•∏ Question 10
Í∞Å jobÏóê Îî∞Îùº divorced/married Ïù∏ÏõêÏùò ÎπÑÏú®ÏùÑ ÌôïÏù∏ ÌñàÏùÑ Îïå Í∑∏ Í∞íÏù¥ Í∞ÄÏû• ÎÜíÏùÄ Í∞íÏùÄ?

```python
target = df.groupby(['job','marital']).size().reset_index()
pivotdf = target.pivot_table(index='job', columns = 'marital')[0]
pivotdf = pivotdf.fillna(0) # Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨
pivotdf['ratio'] = pivotdf['divorced'] / pivotdf['married']

result = pivotdf.sort_values('ratio').ratio.values[-1]
print(result)
```

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv')
submission= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/submission.csv')

display(train.head())
display(test.head())
display(submission.head())
```

### Í∞ÑÎã®Ìïú Î™®Îç∏ÎßÅ ÏûëÏóÖ
Î™®Îç∏ÎßÅ Î∞è submissionÌååÏùº ÏÉùÏÑ±ÍπåÏßÄ

```python
# train.info() # Í≤∞Ï∏°Ïπò ÏóÜÏùå
# train.select_dtypes(include = object).columns 
# ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']

# test.info() Í≤∞Ï∏°ÏπòÏóÜÏùå
train.shape, test.shape, submission.shape
```

```python
train.describe()
```

```python
from sklearn.model_selection import train_test_split

X = train.drop(columns = ['ID', 'y'])
y = train.y
# job, month : nunique 12
Xd = pd.get_dummies(X) # ÏûêÎèôÏúºÎ°ú Î≤îÏ£ºÌòï Î≥ÄÏàò(object) Îßå Ïù∏ÏΩîÎî© ÏàòÌñâ

X_train, X_test, y_train, y_test = train_test_split(Xd, y, stratify=y, random_state=42 )

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
prob = rf.predict_proba(X_test)

from sklearn.metrics import roc_auc_score, classification_report
print('test roc score : ', roc_auc_score(y_test, prob[:,1]))
# print(prob)

Xt = test.drop(columns = ['ID'])
y_pred = rf.predict_proba(pd.get_dummies(Xt))
submission['predict'] = y_pred[:,1]

print('submission file - ')
display(submission.head())
submission.to_csv('004001561.csv', index=False)
```
