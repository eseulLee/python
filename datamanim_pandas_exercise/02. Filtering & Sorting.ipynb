### Q33. df2 의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터프레임화 한 후, item_name 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남기기 ★

df.drop_duplicates( col_nm )

### Q34. df2 의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터프레임화 한 후, item_name 기준으로 중복행 있으면 제거하되 마지막 케이스 남기기 ★

df.drop_duplicates( col_nm , keep = 'last' )

### Q35. df2 데이터 중 new_price 값이 new_price 값의 평균값 이상을 가지는 데이터 인덱싱

### Q38. df2 데이터 중 choice_description 값이 NaN 인 데이터를 NoData 값으로 대체 (loc 이용)

### Q39. df2 데이터 중 choice_description 값에 Black 이 들어가는 경우 인덱싱

### Q40. df2 데이터 중 choice_description 값에 Vegetables 들어가지 않는 경우의 개수 출력

- 개수 : len()
- 들어가는 > 들어가지 않는 : ~ (not) 사용

```python
Ans40 = len(df2.loc[~df2.choice_description.str.contains('Vegetables')])
Ans40
```
