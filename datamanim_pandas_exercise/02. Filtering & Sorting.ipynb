### Q20. 데이터 로드

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'

df2 = pd.read_csv(DataUrl)
Ans20 = type(df2)
Ans20
```

### Q21. quantity 컬럼 값이 3인 데이터를 추출하여 첫 5행 출력

```python
Ans21 = df2[df2.quantity == 3].head(5)
Ans21
```

### Q22. quantity 컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행 출력 ★

```python
Ans22 = df2[df2.quantity == 3].reset_index(drop = True)
Ans22
```

### Q23. quantity, item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임 정의

```python
Ans23_df =  df2[['quantity', 'item_price']]
Ans23_df
```

### Q24. item_price 컬럼의 달러표시 문자 제거, float 타입으로 저장 후 new_price컬럼에 저장 ★

```python
df2['new_price'] = df2['item_price'].str[1:].astype('float')
Ans24 = df2['new_price'].head()
Ans24
```

### Q25. new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수 세기

```python
Ans25 = len(df2[df2['new_price'] <= 5])
Ans25
```

### Q26. item_name 명이 Chicken Salad Bowl 인 데이터 프레임을 추출하라고 index값을 초기화

```python
Ans26 = df2[df2['item_name'] == 'Chicken Salad Bowl'].reset_index(drop=True)

Ans26
```

### Q27. new_price 값이 9 이하이고 item_name 이 Chicken Salad Bowl 인 데이터프레임 추출

```python
Ans27 = df2[(df2.item_name == 'Chicken Salad Bowl') & (df2.new_price <= 9)]
Ans27
```

### Q28. df2 의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화

```python
Ans28 = df2.sort_values('new_price', ascending=False).reset_index(drop=True)
Ans28.head(4)
```

### Q29. df2의 item_name 칼럼 중 Chips포함하는 경우의 데이터 출력

```python
Ans29 = df2[df2.item_name.str.contains('Chips')]
Ans29.head()
```
