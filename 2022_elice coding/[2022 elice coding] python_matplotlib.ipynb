```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## üìå Line Plot

```python
fig, ax = plt.subplots()  # subplot() Í¥ÑÌò∏ ÏïàÏóê ÏïÑÎ¨¥Í≤ÉÎèÑ ÏóÜÏúºÎ©¥ 1Í∞úÏùò Í∑∏ÎûòÌîÑ ÏÉùÏÑ±
x = np.arange(15)
y = x ** 2
ax.plot(
    x, y,
    linestyle = ':',
    marker = '*',
    color = '#524FA1'
)
```

### üí° Line Style

```python
x = np.arange(10)
fig, ax = plt.subplots()

# solid
ax.plot(x, x, linestyle = '-')

# dashed
ax.plot(x, x+2, linestyle = '--')

# dastdot
ax.plot(x, x+4, linestyle = '-.')

# dotted
ax.plot(x, x+6, linestyle = ':')
```

### üí° Line Color

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, color = 'r')  # red
ax.plot(x, x+2, color = 'green')
ax.plot(x, x+4, color = '0.8')  # 0~1 ÏÇ¨Ïù¥Ïùò grayscale
ax.plot(x, x+6, color = '#524FA1')
```

### üí° Marker



```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, marker = '.')
ax.plot(x, x+2, marker = 'o')
ax.plot(x, x+4, marker = 'v')
ax.plot(x, x+6, marker = 's')  # square
ax.plot(x, x+8, marker = '*')
```

### üí° Ï∂ï Í≤ΩÍ≥Ñ Ï°∞Ï†ïÌïòÍ∏∞

```python
x = np.linspace(0, 10, 1000)  #  np.linspace(start, end, step)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-2, 12)    # x limit
ax.set_ylim(-1.5, 1.5) # y limit
```

### üí° Î≤îÎ°Ä

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, label = 'y=x')
ax.plot(x, x**2, label = 'y=x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='upper right',
         shadow=True,
         fancybox=True,
         borderpad=2)
```

## üìå Bar Plot

```python
x = np.arange(10)
fig, ax = plt.subplots(figsize=(12, 4))  # (Í∞ÄÎ°ú, ÏÑ∏Î°ú)
ax.bar(x, x*2)  # (x, y)
```

```python
x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)
data = [x, y, z]

fig, ax = plt.subplots()
x_ax = np.arange(3)
for i in x_ax :
    ax.bar(x_ax, data[i],
    bottom = np.sum(data[:i], axis=0))

ax.set_xticks(x_ax)
ax.set_xticklabels(['A', "B", "C"])
```

## üìå Histogram (ÎèÑÏàòÎ∂ÑÌè¨Ìëú)

```python
fig, ax = plt.subplots()
data = np.random.randn(1000)
ax.hist(data, bins=50)   # bins : ÎßâÎåÄÏùò Í∞úÏàò
```

## ü•∏ Î≤îÎ°Ä(legend) ÏúÑÏπò Î≥ÄÍ≤Ω


```python
# legend ÏúÑÏπòÎ•º "ÏôºÏ™Ω Ï§ëÍ∞Ñ"ÏúºÎ°ú ÏÑ∏ÌåÖÌï¥Î≥¥Í∏∞
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(
    x, x, label = 'y=x',
    linestyle = '-',
    marker = '.',
    color = 'blue'
)
ax.plot(
    x, x**2, label= 'y=x^2',
    linestyle = '-.', 
    marker = ',',
    color = 'red'
)
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.legend(
    loc = 'center left',  # 'upper left'
    shadow=True,
    fancybox=True,
    borderpad=2
)
```
